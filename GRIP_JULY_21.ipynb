###**Importing Essential Libraries**

```python
import keras_ocr as kr
import matplotlib.pyplot as plt
```

```python
# Automatically downloads pretrained weights for the detector and recognizer
pipeline=kr.pipeline.Pipeline()
```

###**Importing Images**

```python
image_ocr=[
           kr.tools.read(url) for url in['https://user-images.githubusercontent.com/25033310/74313656-fece6500-4d6b-11ea-8fae-082421897241.jpg',
                                        'https://i0.wp.com/bintransition.com/wp-content/uploads/2019/06/img_0814.jpg?ssl=1',
                                        'https://upload.wikimedia.org/wikipedia/commons/b/b4/EUBanana-500x112.jpg'

                                         ]
]
```

```python
len(image_ocr)
```

###**Displaying the Image**

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[0])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[1])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[2])
```

```python
prediction_groups=pipeline.recognize(image_ocr)
```

**Get predictions**

```python
fig, axs = plt.subplots(nrows=len(image_ocr), figsize=(15, 15))
for ax, image, predictions in zip(axs, image_ocr, prediction_groups):
    kr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
